<H1>Visualize a single author-pair</H1>

```r
two_dtc_stats = function(dtc1, dtc2) {
    author_stat = function(dtc){
    dtc %>% ungroup() %>% droplevels() %>%
    select(doc_id, term, n) %>%
    complete(doc_id, term, fill = list(n = 0)) %>%
     mutate(no_docs = n_distinct(dtc$doc_id)) %>%
     group_by(doc_id) %>%
     mutate(T.in_doc = sum(n), r = 2*sqrt(n + 2/8) / sqrt(T.in_doc)) %>%
     ungroup() %>%
     group_by(term) %>%
     group_by(term, no_docs) %>%
     summarize(mu = mean(r), std = sqrt(var(r)), evar = mean(1/T.in_doc), n = sum(n)) %>%
     select(term, no_docs, mu, std, n, evar) %>%
     unique()
    }
    
    anscomb1 <- author_stat(dtc1) %>% filter(n > 2)
    anscomb2 <- author_stat(dtc2) %>% filter(n > 2)

    res <- anscomb1 %>% full_join(anscomb2, by = 'term') %>%
            ungroup() %>%
            fill(no_docs.x, no_docs.y) %>%
            replace(., is.na(.), 0) %>%
            mutate(var_pooled = ((no_docs.x-1) * std.x^2 + (no_docs.y-1) * std.y^2) / (no_docs.x + no_docs.y-1),
              z_p = (mu.y - mu.x) / sqrt(var_pooled),
              z1 = (mu.y - mu.x) / std.x,
              z2 = (mu.x - mu.y) / std.y,
              tot.x = sum(n.x),
              tot.y = sum(n.y),
              ) %>%
            rowwise() %>%
            mutate(p = ifelse(n.x + n.y > 0, 
                   binom.test(x = n.x, n = n.x+n.y, p = (tot.x-n.x) / (tot.x + tot.y - n.x - n.y))$p.value,
                   1)) %>%
            arrange(abs(p)) %>%
            mutate(rank = order(p)) %>%
            ungroup()
    res
}    
```

```r
#sample two corpora
lo_classes = unique(dtc$class)

df_smp <- expand.grid(lo_classes, lo_classes) %>% 
          filter(as.vector(Var2) > as.vector(Var1)) %>%
          sample_n(1)

smp = df_smp[1,]

#smp = data_frame(Var1 = c('p05-c01'), Var2 = c('p05-c06')) # Harry potter

#smp$Var1 = 'p02-c08'
#smp$Var2 = 'p04-c09'

dtc1 = dtc %>% filter(class == smp$Var1)
dtc2 = dtc %>% filter(class == smp$Var2)

lo_doc_id = unique(dtc1$doc_id)
lo_doc_id2 = unique(dtc2$doc_id)

print(length(lo_doc_id))
print(length(lo_doc_id2))
```

```r
top_rank = 1000
k = 3

dc = sample(lo_doc_id, k)
dc2 = sample(lo_doc_id2, k)

dtc_corpus = dtc1 %>% filter(!(doc_id %in% dc) )
dtc_doc = dtc1 %>% filter(doc_id %in% dc)
dtc_other_corpus = dtc2 %>% filter(!(doc_id %in% dc2) )

stat_within <- two_dtc_stats(dtc_doc, dtc_corpus)
stat_between <- two_dtc_stats(dtc_doc, dtc_other_corpus)

res0 <- stat_within %>%
        filter(rank < top_rank) %>%
        filter(n.y > 4) %>%
        arrange(abs(p)) %>%
        mutate(rank = order(p),
               type = 'concordant',
               nstd = std.y / mu.y) %>%
        select(rank,p,type,nstd,
               z = z1, z_p = z_p,
               std = std.y,
               mu = mu.y,
              evar = evar.y) 

p_th = hc.vals(res0$p, stbl = FALSE, alpha = 0.25)$p.star
n_th = sum(res0$p <= p_th)
res0 <- res0 %>% mutate(n_th = n_th, thresh = rank < n_th) 

res1 <- stat_between %>%
        filter(rank < top_rank) %>%
        filter(n.y > 4) %>%
        arrange(abs(p)) %>%
        mutate(rank = order(p),
               type = 'discordant',
               nstd = std.y / mu.y) %>%
        select(rank, type, p, nstd,
               z = z1,
               z_p = z_p,
               std = std.y,
               mu = mu.y,
              evar = evar.y) 

p_th = hc.vals(res1$p, stbl = FALSE, alpha = 0.25)$p.star
n_th = sum(res1$p <= p_th)

res1 <- res1 %>% mutate(n_th = n_th, thresh = rank < n_th)


#between two corpus
res <- two_dtc_stats(dtc1,dtc2) #%>% filter(n.x + n.y > 3)
p_th = hc.vals(res$p, stbl = FALSE, alpha = 0.25)$p.star
n_th = sum(res$p <= p_th)

res <- res %>% mutate(n_th = n_th)
```

```r
rec <- stat_within %>% select(term, n.x, n.y, tot.y, tot.x) %>%
                left_join(stat_between, by = 'term') %>%
                select(term, n.y.x, n.y.y, n.x.x) %>%
                filter(n.y.x >0, n.y.y > 0) %>%
                remove_missing %>%
                transmute(term = term, n.A = n.y.x , n.B = n.y.y, n = n.x.x,
                          tot.A= sum(n.A), tot.B = sum(n.B), tot = sum(n)) %>%
                arrange(-(n.A + n.B)) 

rec %>% 
        mutate(f.A = n.A/tot.A, f.B = n.B/tot.B, f = n / tot,
               chi.A = (f * sqrt(tot.A / tot) - f.A * sqrt(tot/tot.A) )^2 / (f.A + f),
               chi.B = (f * sqrt(tot.B / tot) - f.B * sqrt(tot/tot.B) )^2 / (f.B + f),
              ) %>%
        summarize(chi.A = sum(chi.A), chi.B = sum(chi.B))
```
